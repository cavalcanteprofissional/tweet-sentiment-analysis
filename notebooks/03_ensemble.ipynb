# 03 - Ensemble (Análise Final)

Combinação dos 3 modelos via weighted soft voting.

```python
from datasets import load_dataset
import pandas as pd
from src.models import load_pipelines, LABEL_MAP
from src.preprocessor import clean_tweet
from src.ensemble import ensemble_predict, DEFAULT_WEIGHTS, batch_predict
from src.evaluate import full_report, compare_models
from tqdm import tqdm

ds = load_dataset('cardiffnlp/tweet_eval', 'sentiment')
df_test = pd.DataFrame(ds['test'])
```

## Avaliar Ensemble

```python
pipes = load_pipelines()

preds = []
for text in tqdm(df_test['text']):
    pred, _ = ensemble_predict(clean_tweet(text), pipes, DEFAULT_WEIGHTS)[:2]
    preds.append(pred)

full_report(df_test['label'], preds, 'Ensemble')
```

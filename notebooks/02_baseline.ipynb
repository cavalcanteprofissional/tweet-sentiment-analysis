# 02 - Modelos Individuais (Baseline)

Avaliação de cada modelo individualmente.

```python
from datasets import load_dataset
import pandas as pd
from src.models import load_pipelines, LABEL_MAP
from src.preprocessor import clean_tweet
from src.evaluate import full_report
from tqdm import tqdm

ds = load_dataset('cardiffnlp/tweet_eval', 'sentiment')
df_val = pd.DataFrame(ds['validation'])
```

## Carregar Pipelines

```python
pipes = load_pipelines()
print('Pipelines carregados!')
```

## Avaliar RoBERTa Twitter

```python
from src.models import predict_single

preds = []
for text in tqdm(df_val['text']):
    pred, _ = predict_single(clean_tweet(text), pipes['roberta_twitter'], 
                         LABEL_MAP['roberta_twitter'])
    preds.append(pred)

full_report(df_val['label'], preds, 'RoBERTa Twitter')
```

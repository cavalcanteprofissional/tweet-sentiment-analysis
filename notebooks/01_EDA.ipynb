# 01 - EDA (Análise Exploratória)

Análise exploratória do dataset tweet_eval.

```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## Carregamento do Dataset

```python
ds = load_dataset('cardiffnlp/tweet_eval', 'sentiment')
print(ds)
```

## Distribuição de Classes

```python
df_train = pd.DataFrame(ds['train'])
df_train['label'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição de Classes - Train')
plt.xlabel('Label')
plt.ylabel('Count')
plt.show()
```

## Exemplos por Classe

```python
label_names = {0: 'negative', 1: 'neutral', 2: 'positive'}
for label in [0, 1, 2]:
    print(f"\n=== {label_names[label].upper()} ===")
    samples = df_train[df_train['label'] == label]['text'].head(3)
    for s in samples:
        print(f"- {s[:100]}...")
```
